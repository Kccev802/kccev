## **1. Human Visual Perception: Mach Bands**
**Exercise 1.1:** Simulating Mach bands to show how our eyes perceive transitions.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create Mach Bands
bands = np.zeros((100, 600))
for i in range(6):
    bands[:, i*100 : (i+1)*100] = i*40

plt.imshow(bands, cmap='gray')
plt.title("Synthetic Mach Bands")
plt.axis('off')
plt.show()
```

## **2. Simultaneous Contrast**
**Exercise 2.1:** Perceived brightness depends on the background.

```python
img = np.ones((200, 600)) * 100
img[:, :200] = 20
img[:, 400:] = 200
img[50:150, 250:350] = 120

plt.imshow(img, cmap='gray')
plt.title("Simultaneous Contrast Example")
plt.axis('off')
plt.show()
```

## **3. Sampling (Spatial Resolution)**
**Exercise 4.1:** Reducing the number of pixels in an image.

```python
# Create a random image for testing
image = np.random.randint(0, 256, (256, 256))

# Downsampling
down_64 = image[::4, ::4]
down_32 = image[::8, ::8]

plt.figure(figsize=(9,3))
plt.subplot(1,3,1); plt.imshow(image, cmap='gray'); plt.title("256x256")
plt.subplot(1,3,2); plt.imshow(down_64, cmap='gray'); plt.title("64x64")
plt.subplot(1,3,3); plt.imshow(down_32, cmap='gray'); plt.title("32x32")
plt.show()
```

## **4. Quantisation (Intensity Resolution)**
**Exercise 5.1:** Reducing the number of grey levels.

```python
def quantise(img, levels):
    step = 256 // levels
    return (img // step) * step

q16 = quantise(image, 16)
q4 = quantise(image, 4)

plt.figure(figsize=(6,3))
plt.subplot(1,2,1); plt.imshow(q16, cmap='gray'); plt.title("16 Levels")
plt.subplot(1,2,2); plt.imshow(q4, cmap='gray'); plt.title("4 Levels")
plt.show()
```

# Setup & Original Imag

```python
import numpy as np
import matplotlib.pyplot as plt

x, y = np.meshgrid(np.linspace(-1, 1, 256), np.linspace(-1, 1, 256))
image = np.sin(10 * (x**2 + y**2)) * 127 + 128
image = image.astype(np.uint8)

plt.imshow(image, cmap='gray')
plt.title("Original Image (256x256)")
plt.axis('off')
plt.show()
```

Here we import libraries and make a greyscale image. Im using a math formula to make a fake image with curves so its easy to see the effects of sampling and quantisation.

# Stage 1 - Spatial Resolution (Sampling)

Sampling is the number of pixels in the image. Here I will make 3 versions of the orginal image with lower spatial resolution by taking every 2nd, 4th, and 8th pixel.

```python
res_128 = image[::2, ::2]
res_64 = image[::4, ::4]
res_32 = image[::8, ::8]

plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(res_128, cmap='gray')
plt.title("Resolution: 128x128")

plt.subplot(1, 3, 2)
plt.imshow(res_64, cmap='gray')
plt.title("Resolution: 64x64")

plt.subplot(1, 3, 3)
plt.imshow(res_32, cmap='gray')
plt.title("Resolution: 32x32")

plt.tight_layout()
plt.show()
```

# Stage 2 - Intensity Resolution (Quantisation)

Quantisation decides how many gray levels we have. I used the function from the lab to reduce the levels to 32, 8 and 2.

```python
def quantise(img, levels):
    step = 256 // levels
    return (img // step) * step

quant_32 = quantise(image, 32)
quant_8 = quantise(image, 8)
quant_2 = quantise(image, 2)

plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(quant_32, cmap='gray')
plt.title("Levels: 32")

plt.subplot(1, 3, 2)
plt.imshow(quant_8, cmap='gray')
plt.title("Levels: 8")

plt.subplot(1, 3, 3)
plt.imshow(quant_2, cmap='gray')
plt.title("Levels: 2 (Binary)")

plt.tight_layout()
plt.show()
```

# Degradation Commentary & Reflection Questions

***Commentary on Degradation***
Decreasing sampling (spatial res) hurts the image more. When we drop pixels, the shapes are lost and it becomes very blocky. Decreasing quantisation just makes bands (false contours) but we can still see the overal shape. Human visual preception uses edges so we can understand low colors, but not low pixels.


***Reflection Questions***

1. Why is a digital image always an approximation of reality?

The real world is continuous but a digital image makes it into a fixed grid of pixels (sampling) and fixed light levels (quantisation), so its just an approximation.

2. How does human perception complicate objective image analysis?

Human preception depends on local contrast like the lab showed. What the computer sees in numbers is diffrent from what our eyes see.

3. Why is interpolation unavoidable in geometric transformations?

 When doing geometric transformations, the new pixel places usually fall on fractional numbers. We need interpolation to guess the new pixel value based on its neigbors.
