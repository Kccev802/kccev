```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
image_path = '/content/قلب.jpg'
color_image = np.array(Image.open(image_path))

if len(color_image.shape) == 3:
    gray = color_image.mean(axis=2)
else:
    gray = color_image

min_val = gray.min()
max_val = gray.max()
enhanced = (gray - min_val) / (max_val - min_val) * 255

noise = np.random.normal(0, 20, gray.shape)
noisy = gray + noise

def mean_filter(img, k=3):
    pad = k // 2
    padded = np.pad(img, pad, mode='edge')
    output = np.zeros_like(img)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            output[i, j] = padded[i:i+k, j:j+k].mean()
    return output

denoised = mean_filter(noisy)

threshold = gray.mean()
binary = gray > threshold


plt.figure(figsize=(15, 4))
plt.subplot(1, 4, 1), plt.imshow(gray, cmap='gray'), plt.title("Original")
plt.subplot(1, 4, 2), plt.imshow(noisy, cmap='gray'), plt.title("Noisy")
plt.subplot(1, 4, 3), plt.imshow(denoised, cmap='gray'), plt.title("Denoised")
plt.subplot(1, 4, 4), plt.imshow(binary, cmap='gray'), plt.title("Segmented")
plt.show()
```

```python
#mini project
proj_img = np.zeros((100, 100))
proj_img[20:40, 20:40] = 150 # Object 1
proj_img[60:90, 60:80] = 200 # Object 2
proj_noise = np.random.normal(0, 15, proj_img.shape)
proj_noisy = proj_img + proj_noise
proj_denoised = mean_filter(proj_noisy)
min_v = proj_denoised.min()
max_v = proj_denoised.max()
proj_enhanced = (proj_denoised - min_v) / (max_v - min_v) * 255
proj_binary = proj_enhanced > proj_enhanced.mean()
plt.figure(figsize=(15, 4))
plt.subplot(1, 4, 1), plt.imshow(proj_img, cmap='gray'), plt.title("Original")
plt.subplot(1, 4, 2), plt.imshow(proj_noisy, cmap='gray'), plt.title("Noisy")
plt.subplot(1, 4, 3), plt.imshow(proj_enhanced, cmap='gray'), plt.title("Enhanced")
plt.subplot(1, 4, 4), plt.imshow(proj_binary, cmap='gray'), plt.title("Segmented")
plt.show()
```
